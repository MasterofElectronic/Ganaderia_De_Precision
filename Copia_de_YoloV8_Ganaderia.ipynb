```python
from ultralytics import YOLO

# Carga un modelo YOLOv8n
model = YOLO('yolov8n.pt')

# Muestra información del modelo (opcional)
model.info()

# Entrena el modelo en el conjunto de datos de ejemplo COCO8 durante 100 épocas
results = model.train(data='/content/cowsV82/cows2v8/configuration.yaml', epochs=10, imgsz=640)

```

```python
# Realiza inferencia con el modelo YOLOv8n en la imagen '050c461d-762a-4d53-8c1f-5985affefc71.jpg'
prueba1 = model('/content/cowsV82/cows2v8/test/images/0bfaae43-5e9c-4424-a6b2-21cb6340f760.JPG')
print(prueba1)
```

```python
# Validate the model
metrics = model.val()  # no arguments needed, dataset and settings remembered
metrics.box.map    # map50-95
metrics.box.map50  # map50
metrics.box.map75  # map75
metrics.box.maps   # a list contains map50-95 of each category
```
